```python
# import relevant packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV

from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.svm import SVC

from sklearn.metrics import log_loss, roc_auc_score, precision_score, recall_score, confusion_matrix
from scikitplot.metrics import plot_precision_recall, plot_roc

# set display options for the notebook

pd.options.display.max_columns = None
sns.set_style("darkgrid")

# set up data directory

dirData = "data/"

# define data file name

fileTrain = 'denorm_data.csv'
fileHoldOut = 'denorm_holdout.csv'
```

```python
# import denormalized training dataset with all the available features to be used in model training and testing

train = pd.read_csv("{}/{}".format(dirData, fileTrain))

# import denormalized hold-out dataset to test the algorithms on unseen dataset

holdout = pd.read_csv("{}/{}".format(dirData, fileHoldOut))
```

```python
# number of users in training dataset

train['msno'].count()
```

```python
# number of users in hold-out dataset

holdout['msno'].count()
```

```python
# number of CPU cores in the machine
# will be helpful in instantiating n_jobs parameter while training algorithms later

import os
cores = np.int(os.cpu_count()/2)

print("Total CPU cores on the machine:", cores*2)
print("Number of cores to be used:", cores)
```

```python
# function to prepare x and y datasets

def prepXY(data, features='Null'):
    
    # limit the attributes to selected features
    if features=='Null':
        df = data.copy()
    else:
        df = data[features].copy()

    # split data into x and y
    x = df.set_index('msno').drop('churn', axis=1)
    y = df.set_index('msno')['churn']
    
    # scale or normalize the features to a given range of 0 and 1
    scaler = MinMaxScaler(feature_range=(0, 1))
    x_scaled = pd.DataFrame(data=scaler.fit_transform(x), columns=x.columns, index=x.index)
    
    return(x_scaled, y)

# function to search the grid of paramters to find the best ones

def gridSearch(clf, params, x, y):
    
    # create Grid Search object
    gs = GridSearchCV(estimator=clf, param_grid=params, scoring='neg_log_loss', n_jobs=cores,
                      cv=StratifiedKFold(n_splits=5, random_state=101), refit=True, return_train_score=False)
    
    gs.fit(x, y)
    
    print("Results of training and tuning the model -\n")
    print("Best hyper-parameters: {}".format(gs.best_params_))
    print("Best log-loss score: {}".format(gs.best_score_))
    
    return gs

# function to calculate and print relevant scores of the trained or tested model

def calcScores(clf, x, y):
    
    # predict y from test data
    y_pred = clf.predict(x)
    
    # predict probability of y from test data
    y_pred_prob = clf.predict_proba(x)
    
    logloss = round(log_loss(y, y_pred_prob[:,1]), 4)
    baselogloss = round(log_loss(y, np.full(y.shape, y.value_counts(normalize=True)[1])), 3)
    rocauc = round(roc_auc_score(y, y_pred_prob[:,1]), 3) 
    prc = round(precision_score(y, y_pred) * 100, 1)
    rcl = round(recall_score(y, y_pred) * 100, 1)
    cnf = confusion_matrix(y, y_pred)

    print("\nResults of testing the model -")

    print('\nLog-Loss:', logloss, "\tBase Log-Loss:", baselogloss)
    print('Precision:', prc, '\tSensitivity: ', rcl)
    print('ROC AUC Score:', rocauc)
    print('\nConfusion Matrix: \n', cnf)
    
#    print("\n ROC Curve -")
    plot_roc(y, y_pred_prob, classes_to_plot=[1], plot_micro=False, plot_macro=False)

#    print("\n Precision-Recall Curve -")
    plot_precision_recall(y, y_pred_prob, classes_to_plot=[1], plot_micro=False)
```

```python
def plotLogLossFixedProb(yTrue):
    
    # plot impact of logloss for single forecasts
    from sklearn.metrics import log_loss
    import matplotlib.pyplot as plt
    from numpy import array

    # probability predictions as 0 to 1 in 0.01 increments
    prob = [i*0.001 for i in range(0, 1001)]

    # compute log_loss values for each value of probability fixed for the entire dataset
    losses = [log_loss(yTrue, np.ones(yTrue.shape)*p, labels=[0,1]) for p in prob]        
    minLogLoss = round(min(losses), 3)
    
    ymin = min(losses)
    xmin = prob[losses.index(ymin)]

    print('Balance in the dataset:\n', round(yTrue.value_counts(normalize=True)*100, 2))
    
    # plot input to loss
    plt.plot(prob, losses, label='LogLoss')
    plt.title('Evaluating model with constant probability')
    plt.xlabel('Predicted Probability')
    plt.ylabel('Log Loss')
    plt.annotate('Local min: (base log-loss of ' + str(minLogLoss) + '\nat probability ' + str(prob[losses.index(ymin)]) + ')',
                 xy=(xmin, ymin), xytext=(xmin, ymin+10),
                arrowprops=dict(facecolor='black', shrink=0.05),
                )
    plt.annotate('Distribution (in %):\n' + 'Class 0: ' + str(round(yTrue.value_counts(normalize=True)*100, 2)[0]) + '\nClass 1: ' + str(round(yTrue.value_counts(normalize=True)*100, 2)[1]),
                 xy=(0.1, max(losses)-5), xytext=(0.1, max(losses)-5))
    
    plt.show()
```

```python
## Prepare training dataset for model fitting and tuning

x, y = prepXY(train)

print("Churn distribution in train dataset:\n", y.value_counts(normalize=True))
```

```python
## Prepare hold-out dataset for eventual testing

hx, hy = prepXY(holdout)

print("Churn distribution in hold-out dataset:\n", hy.value_counts(normalize=True))
```

```python
# base log loss score of the naive model on training set

plotLogLossFixedProb(y)
```

```python
# base log loss score of the naive model on hold-out set

plotLogLossFixedProb(hy)
```

### Extreme Gradient Boosting (XGB)

```python
## Tuning the classifier

# instantiate the classifier
xgb = XGBClassifier(objective='binary:logistic', eval_metric='logloss',
                    booster='gbtree', silent=0, n_jobs=cores, random_state=101)

# define grid of parameters
params = {'n_estimators': [125, 150, 175],
          'max_depth': [3, 5, 7],
          'learning_rate': [0.03, 0.05, 0.07]}

# search the grid for best parameters
gsxgbclf = gridSearch(xgb, params, x, y)

# calculate classification scores on hold-out dataset
print("\n********************")
print("Results on Hold-Out dataset")
calcScores(gsxgbclf, hx, hy)
```

```python
## Feature Importance out of tuned XGB model

# store the importance of features in a dataframe
fimp = pd.DataFrame({'feature': x.columns,
                     'importance': gsxgbclf.best_estimator_.feature_importances_})

# plot horizontal bar chart to show importance of features in the model
fimp.sort_values('importance').plot(x='feature', y='importance', kind='barh', legend=False,
                                                    figsize=(9,len(fimp.columns)*3), title='Feature Importance',
                                                    fontsize=12, color='blue', alpha=0.5)
print('\n', list(fimp['feature']))
```

### Random Forest Classifier

```python
## Tuning the classifier

# instantiate the classifier
rfc = RandomForestClassifier(max_features='auto',
                             n_jobs=cores, random_state=101)

# define grid of parameters
params = {'n_estimators': [200, 250, 300],
          'max_depth': [13, 16, 19],
          'min_samples_split':[10, 13, 16],
          'criterion':['entropy']}

# search the grid for best parameters
gsrfclf = gridSearch(rfc, params, x, y)

# calculate classification scores on hold-out dataset
print("\n********************")
print("Results on Hold-Out dataset")
calcScores(gsrfclf, hx, hy)
```

### Bagging Classifier

```python
## Tuning the classifier

# instantiate the classifier
bgc = BaggingClassifier(bootstrap=True, n_jobs=cores, random_state=101)

# define grid of parameters
params = {'n_estimators': [225, 250, 300],
          'max_features':[13, 15, 17]}

# search the grid for best parameters
gsbgclf = gridSearch(bgc, params, x, y)

# calculate classification scores on hold-out dataset
print("\n********************")
print("Results on Hold-Out dataset")
calcScores(gsbgclf, hx, hy)
```

### Support Vector Classifier (SVC)

```python
%%time

## SVC model training and testing

svc = SVC(gamma='auto', random_state=101, probability=True)

# fit SVC on train dataset
svc.fit(x, y)

# calculate classification scores on hold-out dataset
print("\n********************")
print("Results on Hold-Out dataset")
calcScores(svc, hx, hy)
```
