```python
# import relevant packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# set display options for the notebook

from IPython.display import display
pd.options.display.max_columns = None
sns.set_style("darkgrid")

# set up data directory

dirData = "data/"

# define file names for datasets to be used in EDA

fileDates = 'dates.csv'
fileLastTrx = 'last_trx.csv'
fileCleanReg = 'reg_clean.csv'
fileUsgAggLifetime = 'usg_agg_lifetime.csv'
fileUsgAggPeriods = 'usg_agg_periods.csv'
```

```python
## import all the required datasets

# last sales/subscription transaction data
lstrx = pd.read_csv("{}/{}".format(dirData, fileLastTrx))

# clean registration/membership data
reg = pd.read_csv("{}/{}".format(dirData, fileCleanReg), usecols=['msno','churn','regAge','userAge','city','regSrc'])

# aggregated lifetime usage data
adf = pd.read_csv("{}/{}".format(dirData, fileUsgAggLifetime))

# aggregated limited time-period usage data
dfagg = pd.read_csv("{}/{}".format(dirData, fileUsgAggPeriods))

# dates dimension data
dates = pd.read_csv("{}/{}".format(dirData, fileDates), parse_dates=['date'])
```

```python
# save minimum and maximum dates in scope of usage data

startDate = dates['date'].min()
currDate = dates['date'].max()

print(startDate, currDate)
```

```python
# a demo user used to see sample data along the transformations

demUsr = '3Jyl3B1cP+CYYgpEZRDwq5huDkB9qfNOqq3Lt1JYfXE='
```

```python
# functions to analyze churn rate by the given categorical variable

def churnRateBy(df, attr):
    piv = df.groupby(['churn', attr])['msno'].count().unstack('churn').fillna(0).astype(int)
    piv.columns = ['renew','churn']
    piv['users'] = piv[['renew','churn']].sum(1)
    piv['churnRate'] = (piv['churn'].div(piv['users']) * 100).round(1)
    return(piv)

def churnRateViz(piv, thresh):

    plt.subplot(2,1,1)
    piv[piv['users']>thresh]['churnRate'].plot(kind='bar', color='grey', legend='Churn Rate')
    plt.xlabel(piv.index.name)
    plt.ylabel('Churn Percentage')

    plt.subplot(2,1,2)
    piv[piv['users']>thresh]['renew'].plot(kind='bar', stacked=True, color='green', legend='Renew')
    piv[piv['users']>thresh]['churn'].plot(kind='bar', stacked=True, color='orange', legend='Churn')
    plt.xlabel(piv.index.name)
    plt.ylabel('Number of Users')

    plt.tight_layout()
```

### Transactions

We focus on the last transaction of each user to find insights into his churn/renewal.

```python
lstrx.head(2)
```

```python
# explore churn rate by auto-renewal and cancellation 

piv = lstrx.groupby(['churn','cancel','autoRenew'])['msno'].count().unstack('churn')
piv.columns = ['renew','churn']
piv['churnRate'] = (piv['churn'].div(piv.sum(axis=1), axis=0) * 100).round(1)
piv

# user "cancels" an "auto-renewal" subscription plan only, not a manual-renewal
# if the user has not selected auto-renewal plan, there is 33% chance that he will churn
# if the user is on auto-renewal plan and has not cancelled it, there is 0.4% chance that he will churn
# if the user has just cancelled the auto-renewal plan he was on, there is roughly 86% chance that he will churn
```

```python
# explore churn rate by payment method

piv = churnRateBy(lstrx, 'payMethod')
piv[piv['users']>3000].sort_values('churnRate', ascending=False)

# for payment methods with many users (> 3000), 38 shows particularly higher churn rate than others
```

```python
# visualizing churn rate by payment method

plt.figure(figsize=(10,8))
churnRateViz(piv, 3000)
plt.show()
```

```python
# Out of total sample users, what percentage of users paid through method 38 during the last transaction?

lstrx.query("payMethod==38")['msno'].count()*100/lstrx['msno'].count()
```

```python
# explore churn rate by plan term (in days)

piv = churnRateBy(lstrx, 'planTerm')
print(piv[piv['users']>100].sort_values('churnRate', ascending=False))

# visualizing the churn rate by plan term

plt.figure(figsize=(10,6))
churnRateViz(piv, 100)
plt.show()

# churn rate is higher for any plan term that is not for 30 days
# however, the number of users associated to any such plan is negligible
```

```python
# Out of total sample users, what percentage of users are not on 30 days plan, as determined by the last transaction?

lstrx.query("planTerm!=30")['msno'].count()*100/lstrx['msno'].count()
```

```python
# churn rate by days since last sales transaction (in days)

piv = churnRateBy(lstrx, 'daysLastTrx')
print("Top 10 daysLastTrx by Churn Rate:\n")
print(piv[piv['users']>200].sort_values('churnRate', ascending=False).head(10))

# visualizing the churn rate by days since last transaction

plt.figure(figsize=(10,6))
churnRateViz(piv, 200)
plt.show()

# churn rate is higher for users who had last sales transaction more than 29 days
# however, the number of users associated to such transactions is just 9% of total
```

```python
# Out of total sample users, what percentage of users had last transaction beyond 29 days from now?

(lstrx.query("daysLastTrx>29")['msno'].nunique()*100)/lstrx['msno'].nunique()
```

```python
# visualize churn behavior by number of days since last transaction for 60 days and more

plt.figure(figsize=(18,4))

plt.hist([lstrx.query("churn==0 and daysLastTrx>=60")['daysLastTrx'], 
          lstrx.query("churn==1 and daysLastTrx>=60")['daysLastTrx']], stacked=True, bins=200, label=('Renew','Churn'))
plt.legend()
plt.xlim(60,500)

# beyond 60 days since last transaction, probability of churn seems high
```

#### Analysis of recent vs. lifetime usage behavior

Now that we have usage weekly stats aggregated for different time-periods, we can see change in behavior of churn vs. renew users over time.

Different time periods:

    Lifetime, abbreviated as "Lf", comprises of lifetime usage information of a user
    
    Recent usage is represented by following two types:
    
    Immediate-past, abbreviated as "Ip", consists of usage information for last 2 weeks only
    
    Near-past, abbreviated as "Np", consists of usage information for last 8 weeks only

```python
dfagg.info()
```

```python
# average number of days active per week during lifetime

sns.violinplot(data=dfagg, x='churn', y='daysActivePerWkLf', cut=0, scale='area', gridsize=500)
plt.ylim(0,7)

# churn users might be recording slightly lower activeness over their lifetime than renew users
```

```python
# proportion of lifetime weeks not used

sns.violinplot(data=dfagg, x='churn', y='noUsgPctLf', cut=0, scale='area', gridsize=500)
plt.ylim(-0.00, 1)

# renew users logged into the application for higher proportion of weeks available during their lifetime, than churn users
```

```python
# maximum number of successive weeks an average user stayed away from the service

sns.violinplot(data=dfagg, x='churn', y='noUsgStrkMaxLf', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 35)

# looks like, on average, churn users stayed away from the service longer than renew users, during their lifespan
```

```python
# songs listened to per week during lifetime

sns.violinplot(data=dfagg, x='churn', y='songsPerWkLf', cut=0, scale='area', gridsize=500)
plt.ylim(0,450)

# no difference between churn and renew users in terms of number of songs listened to per week during lifetime
```

```python
## immediate-past behavior

plt.figure(figsize=(16,4))

# number of active days per week in the immediate-past by renew vs. churn user
plt.subplot(1,3,1)
sns.violinplot(data=dfagg, x='churn', y='daysActivePerWkIp', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 4.5)

# number of no-usage weeks in the immediate-past by renew vs. churn user
plt.subplot(1,3,2)
sns.violinplot(data=dfagg, x='churn', y='noUsgWksIp', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 2)

# number of songs per week in the immediate-past by renew vs. churn user
plt.subplot(1,3,3)
sns.violinplot(data=dfagg, x='churn', y='songsPerWkIp', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 300)

# all three metrics differentiate two set of users prominently
```

```python
## near-past behavior

plt.figure(figsize=(16,4))

# number of active days per weeks in the near-past by renew vs. churn user
plt.subplot(1,3,1)
sns.violinplot(data=dfagg, x='churn', y='daysActivePerWkNp', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 6.5)

# number of no-usage weeks in the near-past by renew vs. churn user
plt.subplot(1,3,2)
sns.violinplot(data=dfagg, x='churn', y='noUsgWksNp', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 7)

# number of songs per weeks in the near-past by renew vs. churn user
plt.subplot(1,3,3)
sns.violinplot(data=dfagg, x='churn', y='songsPerWkNp', cut=0, scale='area', gridsize=500)
plt.ylim(-0.05, 350)

# no-usage weeks differentiates two sets of users more prominently that other two metrics
```

```python
## compare usage metrics for immediate past with those for near past

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
sns.violinplot(data=dfagg, x='churn', y='daysActivePerWk%ChgIpNp', cut=0, scale='area', gridsize=500)
plt.ylim(-1.05, 0.75)

plt.subplot(1,2,2)
sns.violinplot(data=dfagg, x='churn', y='songsPerWk%ChgIpNp', cut=0, scale='area', gridsize=500)
plt.ylim(-1.05, 0.75)

# immediate usage of churn users, when compared to near-past usage, declined by a higher degree than that of renew users
```

```python
## compare usage metrics for immediate past with those for lifetime

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
sns.violinplot(data=dfagg, x='churn', y='daysActivePerWk%ChgIpLf', cut=0, scale='area', gridsize=500)
plt.ylim(-1.05, 0.75)

plt.subplot(1,2,2)
sns.violinplot(data=dfagg, x='churn', y='songsPerWk%ChgIpLf', cut=0, scale='area', gridsize=500)
plt.ylim(-1.05, 0.75)

# immediate-past usage of churn users, when compared to lifetime usage, declined by a higher degree than that of renew users
```

```python
## compare usage metrics for near past with those for lifetime

plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
sns.violinplot(data=dfagg, x='churn', y='daysActivePerWk%ChgNpLf', cut=0, scale='area', gridsize=500)
plt.ylim(-1.05, 0.75)

plt.subplot(1,2,2)
sns.violinplot(data=dfagg, x='churn', y='songsPerWk%ChgNpLf', cut=0, scale='area', gridsize=500)
plt.ylim(-1.05, 0.95)

# compared to lifetime usage, usage of churn users in near past declined to the same extent as that of renew users
# it seems reasonable that the users behavior diverged in recent times, just before the user decided to churn
# and hence, near past did not witness that divergence
# the distributions would probably match much more with each other if we excluded 
# last 2-3 weeks of usage (that shows more prominent decline for churn users)
```

```python
# average stats of usage metrics for immediate-past behavior vs. near-past/lifetime behavior

dfagg.groupby('churn')[['daysActivePerWk%ChgIpNp','songsPerWk%ChgIpNp',
                        'daysActivePerWk%ChgIpLf','songsPerWk%ChgIpLf']].mean()
```
